```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("retail_sales_dataset.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.shape)
print("Duplicate rows:",df.duplicated().sum())
```

```python
df["Date"]= pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
gender_counts=df["Gender"].value_counts()
result = pd.DataFrame({
    "Count": gender_counts,
    "Percentage":(gender_counts/gender_counts.sum())*100
})
result
```

```python
product_counts= df["Product Category"].value_counts()
result = pd.DataFrame({
    "Counts": product_counts,
    "Percentage":(product_counts/product_counts.sum())*100
})
result
```

```python
for col in df.select_dtypes(include="object").columns:
    print("\n", col)
    print(df[col].value_counts())
```

```python
df.groupby("Date")["Total Amount"].sum().sort_values(ascending=False).head(10) 
```

```python
df.groupby("Age")["Total Amount"].sum().sort_values(ascending=False).head(10)
```

```python
Total_revenue= df["Total Amount"].sum()
Total_revenue
```

```python
df.groupby("Product Category")["Total Amount"].sum().sort_values().plot(kind="barh")
plt.title("Revenue By Product")
plt.xlabel("Total Amount")
plt.ylabel("Product Category")
plt.show()
```

```python
df.groupby("Gender")["Total Amount"].sum().sort_values().plot(kind="pie", autopct="%1.1f%%")
plt.title("Revenue Contribution by Gender")
plt.ylabel("")
plt.show()
```

```python
df.groupby("Month")["Total Amount"].sum().sort_values().plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.xticks(rotation= 45)
plt.show()
```

```python
df.groupby("Customer ID")["Total Amount"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Customers by Spending")
plt.xlabel("Customer ID")
plt.ylabel("Total Amount")
plt.xticks(rotation= 45)
plt.show()
```

```python
bins = [0, 18, 25, 35, 45, 60, 100]
labels =  ["<18","18-24", "25-34", "35-44", "45-59", "60+"]
df["Age Group"]= pd.cut(df["Age"], bins = bins, labels= labels)
```

```python
df.groupby("Age Group")["Total Amount"].sum().sort_values().plot(kind="bar")
plt.title("Spending by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
plt.show()
```

```python
cross= pd.crosstab(df["Gender"], df["Product Category"])
sns.heatmap(cross, annot= True, cmap="Blues")
plt.title("Gender VS Product Category")
plt.show()
```
